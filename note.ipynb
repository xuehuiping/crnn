```python

import os
import random

folder = '/Users/huihui/data/ocrinaction/Challenge2_Training_Task3_Images_GT/'
names = os.listdir(folder)
names = [name for name in names if name.endswith('.png')]
name = random.choice(names)
print(folder + name)

lines = open(folder+'gt.txt').readlines()
for line in lines:
    if name in line:
        print('\n标注：{}'.format(line))
        
file_name = folder + name

import matplotlib.pyplot as plt
%matplotlib inline

img = plt.imread(file_name)  
plt.imshow(img)  
plt.show()
```
